### **Problem 3: Handling Missing Data and Imputation**  
1. Identify columns with **missing values**.
2. Impute the missing values:
   - For **numerical columns**, replace missing values with the **median**.
   - For **categorical columns**, replace missing values with the **most frequent** value.
3. Justify your choice of imputation methods and check if any missing data remains.

```python
import pandas as pd

df = pd.read_csv("train.csv")

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include='object').columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

missing_values_after = df.isnull().sum()
print("Missing values after imputation:")
print(missing_values_after[missing_values_after > 0])
```

```python
import pandas as pd

df = pd.read_csv("train.csv")

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include='object').columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

missing_values_after = df.isnull().sum()
print("Missing values after imputation:")
print(missing_values_after[missing_values_after > 0])
```

### **Problem 4: Numpy Array Operations**  
1. Convert the numerical columns (such as `LotArea`, `YearBuilt`, `GrLivArea`, `SalePrice`, etc.) into a **NumPy array**.
2. Perform the following NumPy operations on the array:
   - Compute the **mean** and **standard deviation** of each column.
   - Normalize the data using **min-max scaling** and **z-score normalization**.
   - Compute the **dot product** between two numerical columns of your choice.

```python
import numpy as np
# Select only numerical columns
numerical_df = iris_df.select_dtypes(include=[np.number])

# Convert to NumPy array
numerical_array = numerical_df.to_numpy()
```

```python
col_means = np.mean(numerical_array, axis=0)
col_stds = np.std(numerical_array, axis=0)

print("\nMean of each column:\n", col_means)
print("\nStandard Deviation of each column:\n", col_stds)
```

```python
min_vals = numerical_array.min(axis=0)
max_vals = numerical_array.max(axis=0)
min_max_scaled = (numerical_array - min_vals) / (max_vals - min_vals)

print("\nMin-Max Normalized Data (first 5 rows):\n", min_max_scaled[:5])

z_score_scaled = (numerical_array - col_means) / col_stds

print("\nZ-score Normalized Data (first 5 rows):\n", z_score_scaled[:5])
```

```python
lot_area = numerical_array[:, 0]    # assuming LotArea is the first numerical column
sale_price = numerical_array[:, -1] # assuming SalePrice is the last numerical column

dot_product = np.dot(lot_area, sale_price)

print("\nDot Product between LotArea and SalePrice:", dot_product)
```
