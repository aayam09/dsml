**Handling Incorrect Data**

```python
import numpy as np
data = pd.DataFrame({
    "id": [1,2,3,4,5],
    "name" : ["john", "jhang", "jenifer", "jeff", "johan"],
    "age" : [25,-32,19,42,"55"],
    "gender" : ["M","M","Z","F","F"],
    "income" : [400, 700, np.nan, 300, 600]
})
```

```python
data
```

```python
data.age = data.age.astype(int)
```

```python
# sample boundry
filter_age = data[(data.age > 0) & (data.age <= 50)]
filter_age
```

```python
data.age = abs(data.age)
data
```
