```python
titanic_df_2 = pd.read_csv("train.csv")
```

```python
titanic_df_2.groupby("Sex")["Age"].mean()
```

```python
def fill_age(row):
    if pd.isnull(row["Age"]):
        return titanic_df_2[titanic_df_2["Sex"] == row["Sex"]]["Age"].mean()
    else:
        return row["Age"]

titanic_df_2["Age"] = titanic_df_2.apply(fill_age, axis=1)
```

```python
titanic_df_2.isna().sum()
```

```python
titanic_df_2.Age.plot.hist(bins = 20)
```

```python
titanic_df_3 = pd.read_csv("train.csv")
```

```python
mean_ages = titanic_df_3.groupby("Sex")["Age"].transform("mean")
```

```python
titanic_df_3["Age"].fillna(mean_ages,axis=0, inplace=True)
```

```python
titanic_df_3.Age.plot.hist(bins = 20)
```

```python
titanic_df_3.isna().sum()
```
