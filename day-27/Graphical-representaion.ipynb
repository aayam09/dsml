## 1. Bar Graph

```python
titanic_df['Survived'].value_counts().plot(
    kind = "bar",
    title = "Count of survived Passengers",
    figsize = (8,4),
    legend = True,
    yticks = list(range(0,700,50)),
    xlabel = "survived (0=No, 1=Yes)",
    ylabel = "No of Passenger",
    grid = True,
    colormap = "bwr"
    
)
    
```

The figure is a bar chart titled "Count of survived Passengers," showing the number of Titanic passengers who survived (1 = Yes) and did not survive (0 = No). The x-axis labels "survived (0=No, 1=Yes)" indicate the categories, while the y-axis, "No of Passenger," ranges from 0 to 650. The chart reveals approximately 550 passengers did not survive and 340 did, highlighting a significant survival imbalance.

```python
group_by_gender = titanic_df[["Sex","Survived"]].groupby(by ="Sex")
```

```python
pd.crosstab(titanic_df.Survived, titanic_df.Sex).plot(
    kind = "bar",
    title = "Count of survived Passengers",
    figsize = (8,4),
    legend = True,
    yticks = list(range(0,700,50)),
    xlabel = "survived (0=No, 1=Yes)",
    ylabel = "No of Passenger",
    colormap = "bwr"
)
    
```

The figure is a bar chart titled "Count of survived Passengers," showing the number of Titanic passengers who survived (1 = Yes) or did not survive (0 = No) by sex. The x-axis, labeled "survived (0=No, 1=Yes)," indicates survival status, and the y-axis, "No of Passenger," ranges from 0 to 650. Using blue for females and red for males, it displays about 100 female and 450 male non-survivors, and 250 female and 100 male survivors, highlighting a higher female survival rate.

## 4.Scatter Plot

```python
titanic_df["SibSp"] *= 20
titanic_df.plot.scatter(x = "Age", y="Fare")
```

## Scatter Plot: Age vs Fare
This scatter plot shows the relationship between passengers' age and the fare they paid.

Each dot represents a passenger.

The x-axis shows Age, and the y-axis shows Fare.

Most passengers paid fares under $100, regardless of age.

A few passengers aged between 20 and 40 paid very high fares (over $500).

This plot helps us identify trends and outliers in the dataset.
